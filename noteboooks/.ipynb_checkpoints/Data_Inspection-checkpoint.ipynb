```python
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
import cv2
```

# Segmentation label maps

```python
image_path = "../Data/VOC2012/SegmentationClass/2007_000129.png"
```

```python
# PIL reads the label image as it is - with pixels containing integer labels {0,1,2,...,20, 255}
img_pil = Image.open(image_path)   
img_pil_np = np.array(img_pil)
print("PIL original image as np array--")
print("Shape:", img_pil_np.shape)
print("Max:", img_pil_np.max())
fig, [ax1, ax2] = plt.subplots(1,2, figsize=(20,10))
ax1.hist(img_pil_np.flatten(), bins=256)
ax2.imshow(img_pil_np)
plt.show()
print()

# Convert PIL grayscale label image into RGB using its color palette, in the range [0,255]
img_pil_rgb_np = np.array(img_pil.convert('RGB')) 
print("PIL RGB image as np array--")
print("Shape:", img_pil_rgb_np.shape)
print("Max:", img_pil_rgb_np.max())
fig, [ax1, ax2] = plt.subplots(1,2, figsize=(20,10))
ax1.hist(img_pil_rgb_np.flatten(), bins=256)
ax2.imshow(img_pil_rgb_np)
plt.show()
print()

# Plt loads images automatically as RGB, but in the range [0,1]
img_plt = plt.imread(image_path) 
print("plt image--")
print("Shape:", img_plt.shape)
print("Max:", img_plt.max())
fig, [ax1, ax2] = plt.subplots(1,2, figsize=(20,10))
ax1.hist(img_plt.flatten(), bins=256)
ax2.imshow(img_plt)
plt.show()
print()

# OpenCV loads images automatically in color, but as BGR and in range [0,255]
img_cv = cv2.imread(image_path)
print("OpenCV image--")
print("Shape:", img_cv.shape)
print("Max:", img_cv.max())
fig, [ax1, ax2] = plt.subplots(1,2, figsize=(20,10))
ax1.hist(img_cv.flatten(), bins=256)
ax2.imshow(img_cv)
plt.show()
```
